```python
# Imports
import svm
import numpy as np
from sklearn import metrics
import os
import alignment
import pylab as pb
```

```python
#Load Data

l=1 #<---- Choose l

cwd = os.getcwd()+'\..\data\modified_approx_kernel_data';

loopVec = np.arange(0,2400,100);
frobVal = np.zeros(len(loopVec))
timeNew = np.zeros(len(loopVec))
timeOld = np.zeros(len(loopVec))
timeRel = np.zeros(len(loopVec))
for i in range(0,1):
    kerns = os.path.join(cwd,'l{}frobVal_{}.npy'.format(l,i))
    frobVal += np.load(kerns)

    kerns = os.path.join(cwd,'l{}newTime_{}.npy'.format(l,i))
    timeNew += np.load(kerns)

    kerns = os.path.join(cwd,'l{}oldTime_{}.npy'.format(l,i))
    timeOld += np.load(kerns)
    
    kerns = os.path.join(cwd,'l{}relTime_{}.npy'.format(l,i))
    timeRel += np.load(kerns)

frobVal = frobVal/1
timeNew = timeNew/1
timeOld = timeOld/1
timeRel = timeRel/1
```
